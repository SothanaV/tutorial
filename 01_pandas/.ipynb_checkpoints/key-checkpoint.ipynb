```python
#Q3 Show number of customers were born in each month. there are 12 months
```

```python
def getMonth(s):
    return s.split('-')[1]
```

```python
df = df_customer
```

```python
df['month'] = df['dob'].apply(lambda x : getMonth(x))
```

```python
o = df.groupby('month').agg('count')['dob']
print(o)
o = list(o)
print(o)
import matplotlib.pyplot as plt
plt.plot(o)
plt.show()
```
